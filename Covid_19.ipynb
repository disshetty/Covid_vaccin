```python
data = pd.read_csv("cowin_vaccine_data_districtwise.csv")
```

 Forward fill NaN value


```python
data.fillna(data.mean(), inplace=True)
```

 Take the required value for analysis
 Total Sessions Conducted


```python
x = []
y = 7
for i in range(139):
    x.append(y)
    y += 10
```

```python
x
```

 Select the columns


```python
valid_data = data.iloc[:, x]
```

```python
valid_data
```

Droping down unwanted columns

```python
valid_data.drop('Unnamed: 1347', axis=1, inplace=True)
valid_data.drop('Unnamed: 1357', axis=1, inplace=True)
valid_data.drop('Unnamed: 1367', axis=1, inplace=True)
valid_data.drop('Unnamed: 1377', axis=1, inplace=True)
valid_data.drop('Unnamed: 1387', axis=1, inplace=True)
```

Drop the first row


```python
valid_data.drop(0, axis=0, inplace=True)
```

```python
valid_data
```

```python
valid_data = valid_data.dropna()
```

 Split data into independent and dependent variable


```python
X = valid_data.iloc[:, :-1]
y = valid_data.iloc[:, -1]
```

```python
X
```

```python
valid_data
```

Training

```python
from sklearn.model_selection import train_test_split as tts
```

```python
# Split dataset into train and test
X_train, X_test, y_train, y_test = tts(X, y, test_size=0.3, random_state=10)
```

Import

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
```

```python
# Fit the model
linreg.fit(X_train, y_train)
```

 Predict the value

```python

my_pred = linreg.predict(X_test)
```

```python
my_pred
```

Accuracy

```python
my_score = linreg.score(X_test, y_test)
```

```python
print(my_score)
```
